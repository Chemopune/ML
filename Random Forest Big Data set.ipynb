```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
loans = pd.read_csv("lending_club_new_data.csv")
loans.head()
```

```python
loans['good_loans'] = loans['bad_loans'].apply(lambda y: 'yes' if y==0 else 'no')
```

```python
loans.head()
```

```python
X = loans.drop(['bad_loans','good_loans'], axis=1)
y = loans['good_loans']
```

```python
X.head()
```

```python
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=124)
```

```python
print(X_train.shape, X_test.shape)
```

```python
print(y_train.shape, y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model2 = DecisionTreeClassifier()
```

```python
model2.fit(X_train, y_train)
```

```python
prediction2 = model2.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, prediction2))
```

```python
print(classification_report(y_test, prediction2))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model2 = RandomForestClassifier(n_estimators = 150)  # Case sensitive functions
```

```python
rf_model2.fit(X_train, y_train)
```

```python
rf_prediction = model2.predict(X_test)
```

```python
print(confusion_matrix(y_test, rf_prediction))
```

```python
print(confusion_matrix(y_test, prediction2))
```

```python
print(classification_report(y_test, rf_prediction))
```

```python
print(classification_report(y_test, prediction2))
```
